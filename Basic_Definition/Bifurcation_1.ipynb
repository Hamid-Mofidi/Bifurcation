## 1- Saddle-Node Bifurcation:
This occurs when two fixed points of a system, one stable and one unstable, collide and annihilate each other. A simple example is the equation:
$$ f(x; \mu) = x^2 + \mu $$
Here, for $\mu > 0$, there are no fixed points, but for $\mu < 0$, two fixed points emerge, one stable and one unstable.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equation
def x_prime(x, mu):
    return x**2 + mu

# Define the range of mu values
mu_values = np.linspace(-2, 2, 1000)

# Plot x' versus mu
plt.figure(figsize=(8, 4))


# Plot x' versus x
plt.subplot(1, 2, 1)
x_values = np.linspace(-2, 2, 1000)
for mu in np.linspace(-1.5, 1.5, 3):
    plt.plot(x_values, x_prime(x_values, mu), label=f'$\mu = {mu}$')
#plt.scatter(0, x_prime(0, 0), color='red', label='Bifurcation point ($x = 0$)')
plt.xlabel('$x$')
plt.ylabel('$x\'$')
plt.title('Saddle-Node Bifurcation Diagram: $x\' = x^2 + \mu$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid(True)

# Plot x versus mu
plt.subplot(1, 2, 2)
x_values = np.linspace(-2, 2, 1000)
# Calculate mu values using mu = -x^2
mu_values = -x_values**2

plt.plot(mu_values, x_values, label='$\mu = -x^2$')
plt.xlabel('$\mu$')
plt.ylabel('$x$')
plt.scatter(0, 0, color='red', label='Bifurcation point ($\mu = 0$)')
plt.title('Bifurcation Diagram: $ x^2 + \mu = 0$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

## 2- Transcritical Bifurcation:
In this type, the stability of the fixed points changes as they pass through each other. An example is:
$$ f(x; \mu) = \mu x - x^2 $$
At $\mu = 0 $, the fixed points $x = 0 $ and $ x = \mu $ exchange stability.

On the $\mu - x$ plane, from left to right, when $\mu$ increases from negative values to 0 to positive values, the fixed point $x=0$ (resp. $x=\mu$) changes from stable (resp. unstable) to unstable (resp. stable) fixed point.

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the equation
def x_prime(x, mu):
    return mu * x - x**2

# Define the range of mu values
mu_values = np.linspace(-2, 2, 1000)

# Plot x' versus mu
plt.figure(figsize=(8, 4))


# Plot x' versus x
plt.subplot(1, 2, 1)
x_values = np.linspace(-5, 5, 1000)
for mu in np.linspace(-2, 2, 3):
    plt.plot(x_values, x_prime(x_values, mu), label=f'$\mu = {mu}$')
#plt.scatter(0, x_prime(0, 0), color='red', label='Bifurcation point ($x = 0$)')
plt.xlabel('$x$')
plt.ylabel('$x\'$')
plt.title('Transcritical Bifurcation Diagram: $x\' = \mu x - x^2$')
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.legend()
plt.grid(True)

# Plot x versus mu
plt.subplot(1, 2, 2)
x_values = np.linspace(-2, 2, 1000)
# Calculate mu values using mu = -x^2
mu_values = x_values

plt.plot(mu_values, x_values, label='$\mu = x$')
plt.plot( x_values, 0 * x_values, label='$ x = 0$')
plt.xlabel('$\mu$')
plt.ylabel('$x$')
plt.scatter(0, 0, color='red', label='Bifurcation point ($\mu = 0$)')
plt.title('Bifurcation Diagram: $ x = \mu~ and~ x = 0$')
#plt.axhline(0, color='m', linewidth=1.1)
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Logistic map equation
def logistic_map(r, x):
    return r * x * (1 - x)

# Function to generate bifurcation diagram
def bifurcation_diagram(r_values, iterations, transient):
    x_values = []

    # Iterate over each value of r
    for r in r_values:
        x = 0.5  # Initial population size
        # Iterate the logistic map equation
        for _ in range(transient):
            x = logistic_map(r, x)  # Transient period to discard
        for _ in range(iterations):
            x = logistic_map(r, x)
            x_values.append(x)

    return x_values

# Parameters
r_values = np.linspace(2.5, 4.0, 1000)  # Range of r values
iterations = 1000  # Number of iterations per r value
transient = 500  # Number of transient iterations to discard

# Generate bifurcation diagram
bifurcation_data = bifurcation_diagram(r_values, iterations, transient)

# Plot
plt.figure(figsize=(10, 6))
plt.title("Bifurcation Diagram of the Logistic Map")
plt.xlabel("r")
plt.ylabel("Population Size")
plt.ylim(0, 1)
plt.scatter(np.repeat(r_values, iterations), bifurcation_data, s=0.1, c='black', marker='.')
plt.show()

```
